```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd

driver = webdriver.Chrome(executable_path=r"C:\Users\weili\Desktop\Y4S2\BT4222\chromedriver.exe")
url = "https://seekingalpha.com/symbol/TSLA/analysis"
driver.get(url)
time.sleep(2)
```

```python
scroll_pause_time = 1 # You can set your own pause time. My laptop is a bit slow so I use 1 sec
screen_height = driver.execute_script("return window.screen.height;")   # get the screen height of the web
i = 1
scroll_height = 32000

while True:
    # scroll one screen height each time
    driver.execute_script("window.scrollTo(0, {screen_height}*{i});".format(screen_height=screen_height, i=i))  
    i += 1
    time.sleep(scroll_pause_time)
    # update scroll height each time after scrolled, as the scroll height can change after we scrolled the page
    #scroll_height = driver.execute_script("return document.body.scrollHeight;")  
    # Break the loop when the height we need to scroll to is larger than the total scroll height
    if (screen_height) * i > scroll_height:
        break
```

```python
headlines = []

for i in range(357):
    first = "//*[@id='root']/div[1]/main/div[2]/div/div[3]/div/div/div[2]/div/section/div/div[2]/div[2]/article["
    second = "]/div/div/h3/a"
    element = driver.find_element_by_xpath(first + str(i+1) + second).text
    headlines.append(element)

headlines
```

```python
day = []
month = []
date = []
author = []
comments = []
year = []

for i in range(357):
    first = "//*[@id='root']/div[1]/main/div[2]/div/div[3]/div/div/div[2]/div/section/div/div[2]/div[2]/article["
    date_second = "]/div/div/div/span"
    author_second = "]/div/div/div/a[1]"
    
    if (driver.find_element_by_xpath(first + str(i+1) + author_second).text == "Editors' Pick"):
        author_second = "]/div/div/div/a[2]"
        comments_second = "]/div/div/div/a[3]"
    
    else:
        author_second = "]/div/div/div/a[1]"
        comments_second = "]/div/div/div/a[2]"
    
    authorText = driver.find_element_by_xpath(first + str(i+1) + author_second).text
    dateText = driver.find_element_by_xpath(first + str(i+1) + date_second).text
    commentsText = driver.find_element_by_xpath(first + str(i+1) + comments_second).text
    
    if(len(dateText.split(sep = ",")) == 2):
        year.append("2021")
    else:
        year.append(dateText.split(sep = ",")[2])
    
    day.append(dateText.split(sep = ",")[0])
    month.append(dateText.split(sep = ",")[1][1:].split(sep = " ")[0][:3])
    date.append(dateText.split(sep = ",")[1][1:].split(sep = " ")[1])
    author.append(authorText)
    comments.append(commentsText[:-9])
```

```python
import pandas as pd

SA = pd.DataFrame(
    {'Headlines': headlines,
     'Day': day,
     "Month": month,
     'Date': date,
     "Year" : year,
     'No. Comments': comments,
     'Author': author
    })

SA
```
